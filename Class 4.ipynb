```python
# recursive factorial:

def fact(n):
    if n == 0:
        return 1
    return n*fact(n-1)
```

```python
from functools import lru_cache


@lru_cache
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
```

```python
class HashTable:
    def __init__(self, size):
        # Setting up size, slots and data.
        self.size = size
        self.slots = [None] * self.size
        self.data = [None] * self.size

    # We're only using integer values for "key".
    def put(self, key, data):
        # calculating hashvalue
        hashvalue = self.hashfunction(key, len(self.slots))
        print("Storing with hashvalue: %s" % hashvalue)
        # Checking if the slot at that index is empty
        if self.slots[hashvalue] == None:
            self.slots[hashvalue] = key
            self.data[hashvalue] = data
            print("Stored: %s with key: %s" % (key, data))
        else:
            # If key already exists, replace old value
            if self.slots[hashvalue] == key:
                self.data[hashvalue] = data
                print("Stored: %s with key: %s" % (key, data))
            else:
                # Otherwise: find the next available slot
                nextslot = self.rehash(hashvalue, len(self.slots))
                while self.slots[nextslot] != None and self.slots[nextslot] != key:
                    nextslot = self.rehash(nextslot, len(self.slots))
                if self.slots[nextslot] == None:
                    self.slots[nextslot] = key
                    self.data[nextslot] = data
                    print("Stored: %s with key: %s" % (key, data))
                else:
                    self.data[nextslot] = data
                    print("Stored: %s with key: %s" % (key, data))
            
            
    def hashfunction(self, key, size):
        return key%size

    def rehash(self, oldhash, size):
        return (oldhash+1)%size

    def get(self, key):
        startslot = self.hashfunction(key, len(self.slots))
        data = None
        stop = False
        found = False
        position = startslot
        while self.slots[position] != None and not found and not stop:
            if self.slots[position] == key:
                found = True
                data = self.data[position]
                print("Found key at position: %s with data: %s" % (position, data))
            else:
                position = self.rehash(position, len(self.slots))
                if position == startslot:
                    stop = True
        return data

    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)

    def print_table(self):
        for i in range(self.size):
            print(" %s -> %s" % (self.slots[i], self.data[i]))
```

```python
# fibonacci example:

ht = HashTable(99)

def fib(n):
    if n < 2:
        return n
    ans = ht.get(n)
    if not ans:
        ht.put(n, fib(n-1) + fib(n-2))
        ans = ht.get(n)
    return ans

fib(100)
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        current = self.head
        if not current:
            self.head = Node(value)
        else:
            while current.next:
                current = current.next
            current.next = Node(value)

#     def prepend(self, value):
#         current = self.head
#         if not current:
#             self.head = Node(value)
#         else:
#             new_node = Node(value)
#             temp_node = self.head
#             new_node.next = temp_node
#             self.head = new_node

    def insert(self, prev_index, value):
        if prev_index == 0:
            new_node = Node(value)
            current_head = self.head
            new_node.next = current_head
            self.head = new_node
        else:
            current = self.head
            counter = 0
            while prev_index != counter and current:
                prev_node = current
                current = current.next
            if prev_index == counter:
                prev_node.next = Node(value)
                prev_node.next.next = current
                

    def __str__(self):
        current = self.head
        out_str = "[%s" % current.data
        while current.next:
            current = current.next
            out_str += ", %s" % current.data
        out_str += "]"
        return out_str
    
```
